# Building models

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import Pipeline

from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score as r2_score


def rmse(y_true, y_pred, kwargs={}):
    return (mse(y_true, y_pred, **kwargs))**0.5

def scatter_prediction(model, 
                       x, 
                       y_true, 
                       y_scaler=1,
                       title = ""):
    """Evaluates MSE and R^2 of model. 
    Draws scatter plot of (y_true, y_predicted).
    Returns  (mse, r2)
    """
    y_pred = model.predict(x)
    
    if y_scaler!=1:
        y_pred = y_scaler.inverse_transform(y_pred.reshape(1, -1)).ravel()
        y_true = y_scaler.inverse_transform(y_true.reshape(1, -1)).ravel()
    
    model_rmse = rmse(y_true, y_pred)
    model_mae = mae(y_true, y_pred)
    model_r2 = r2_score(y_true, y_pred)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plt.plot(y_true, y_pred, 'o', markersize=2, alpha = 0.2) 
    plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], markersize=2, alpha = 0.5, color='r') 
    ax.set_title(f"RMSE={round(model_rmse, 5)} | MAE={round(model_mae, 5)} | R2={round(model_r2, 5)} | {title}")
    ax.set_xlabel("y_true")
    ax.set_ylabel("y_pred")
    plt.show()
    
    return model_rmse, model_mae, model_r2
```

```python
PR_ION = 'Cu' # 'Cu', 'Ni', 'Pb', 'Al', 'Co', 'Cr', 'NO3'

x = df_norm.drop(['Cu', 'Ni', 'Pb', 'Al', 'Co', 'Cr', 'NO3'], axis=1).to_numpy()
y = df_norm[[PR_ION]].to_numpy()

x.shape, y.shape
```

## 5. Scaling y-data
Train / Validation / Test = 0.7 / 0.2 / 0.1

TRAIN_TEST_RATIO = (Test) / (Train+Validation+Test)

VALIDATION_TRAIN_RATIO = Validation / (Train+Validation)

```python
TEST_TRAIN_RATIO = 0.1
VALIDATION_TRAIN_RATIO = 2 / 9 #7 / 9
RS = None

N_ITER_NO_CHANGE = 20
```

```python
# Scaling y before train_test split in cv only because we aprior know the distribution of ions.
scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y)

y_scaler_multiplier = scaler_y.data_max_[0] - scaler_y.data_min_[0]
print(y_scaler_multiplier)

x_train, x_test, y_scaled_train, y_scaled_test = train_test_split(x, y_scaled, 
                                                    test_size=TEST_TRAIN_RATIO,
                                                    random_state=RS)
scaler_x = StandardScaler()
```

```python
mlp_HIDDEN_LAYER_SIZES = [32]
mlp_ACTIVATION = 'tanh'
mlp_SOLVER = 'adam'
mlp_MAX_ITER = 10000
mlp_TOL = 1e-3
```

## 6. Model - MLP
Ordinary one-hidden-layer with *HIDDEN_LAYER_SIZES* neurons perceptron.
We use MLPRegressor from scikit-learn library.

```python
model = MLPRegressor(hidden_layer_sizes=mlp_HIDDEN_LAYER_SIZES,
                     activation=mlp_ACTIVATION,
                     solver=mlp_SOLVER,
                     early_stopping=True,
                     validation_fraction=VALIDATION_TRAIN_RATIO,
                     n_iter_no_change=N_ITER_NO_CHANGE,
                     learning_rate_init = 0.001,
                     learning_rate='constant',
                     max_iter=mlp_MAX_ITER,
                     random_state=RS,
                     tol=mlp_TOL)

mlp_pipel = Pipeline([('scale', scaler_x), ('mlp', model)])
```

```python
mlp_pipel[1].get_params()
```

```python
mlp_pipel.fit(x_train, y_scaled_train.ravel())
```

```python
scatter_prediction(mlp_pipel, x_train, y_scaled_train, y_scaler = scaler_y, title="Train data")
scatter_prediction(mlp_pipel, x_test, y_scaled_test, y_scaler = scaler_y, title="Train data")
```

## 6.1 Cross-validation 1-hidden-layer MLP varying hidden_neurons.

```python
from sklearn.model_selection import KFold, cross_validate
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score, make_scorer

from tqdm import tqdm

n_splits = 10
scoring = {'rmse': make_scorer(root_mean_squared_error), 
           'mae': make_scorer(mean_absolute_error),
           'r2': make_scorer(r2_score)}

l_hn =  list(range(1, 65)) # [5, 7, 10, 15, 20, 32, 64]

mean_mae, std_mae = [], []
mean_rmse, std_rmse = [], []
mean_r2, std_r2 = [], []

for hn in tqdm(l_hn):
    
    model = MLPRegressor(hidden_layer_sizes=hn,
                         activation=mlp_ACTIVATION,
                         solver=mlp_SOLVER,
                         early_stopping=True,
                         validation_fraction=VALIDATION_TRAIN_RATIO,
                         n_iter_no_change=N_ITER_NO_CHANGE,
                         learning_rate_init = 0.001,
                         learning_rate='constant',
                         max_iter=mlp_MAX_ITER,
                         random_state=RS,
                         tol=mlp_TOL)
    mlp_pipel = Pipeline([('scale', scaler_x), ('mlp', model)])
    
    cv_rf_kf = cross_validate(mlp_pipel, x, y_scaled.ravel(), cv=KFold(shuffle=True, n_splits=n_splits), scoring=scoring)
    
    mean_mae.append(np.mean(cv_rf_kf['test_mae']) * y_scaler_multiplier)
    std_mae.append(np.std(cv_rf_kf['test_mae']) * y_scaler_multiplier)
    
    mean_rmse.append(np.mean(cv_rf_kf['test_rmse']) * y_scaler_multiplier)
    std_rmse.append(np.std(cv_rf_kf['test_rmse']) * y_scaler_multiplier)
    
    mean_r2.append(np.mean(cv_rf_kf['test_r2']))
    std_r2.append(np.std(cv_rf_kf['test_r2']))
```

## 8. Model - KAN


```python
import torch
from kan import KAN

from tools import KAN_es, cross_val_KAN_es, kfold_experiment_es

def scatter_prediction_KAN(model, 
                       x, 
                       y_true, 
                       y_scaler = 1,
                       title = ""):
    """Evaluates MSE and R^2 of model.
    Draws scatter plot of (y_true, y_predicted).
    Returns  (mse, r2)
    """
    y_pred = model.forward(x).detach().numpy()
    y_true = y_true.detach().numpy()
    
    if y_scaler!=1:
        y_pred = y_scaler.inverse_transform(y_pred)
        y_true = y_scaler.inverse_transform(y_true)
    
    model_rmse = rmse(y_true, y_pred)
    model_mae = mae(y_true, y_pred)
    model_r2 = r2_score(y_true, y_pred)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plt.plot(y_true, y_pred, 'o', markersize=2, alpha = 0.2) 
    plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], markersize=2, alpha = 0.5, color='r') 
    ax.set_title(f"RMSE={round(model_rmse, 5)} | MAE={round(model_mae, 5)} | R2={round(model_r2, 5)} | {title}")
    ax.set_xlabel("y_true")
    ax.set_ylabel("y_pred")
    plt.show()
    
    return model_rmse, model_mae, model_r2
```
