# Building models

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neural_network import MLPRegressor

from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score as r2_score


def rmse(y_true, y_pred, kwargs={}):
    return (mse(y_true, y_pred, **kwargs))**0.5

def scatter_prediction(model, 
                       x, 
                       y_true, 
                       title = ""):
    """Evaluates MSE and R^2 of model. 
    Draws scatter plot of (y_true, y_predicted).
    Returns  (mse, r2)
    """
    y_pred = model.predict(x)
    model_rmse = rmse(y_true, y_pred)
    model_mae = mae(y_true, y_pred)
    model_r2 = r2_score(y_true, y_pred)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plt.plot(y_true, y_pred, 'o', markersize=2, alpha = 0.2) 
    plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], markersize=2, alpha = 0.5, color='r') 
    ax.set_title(f"RMSE={round(model_rmse, 5)} | MAE={round(model_mae, 5)} | R2={round(model_r2, 5)} | {title}")
    ax.set_xlabel("y_true")
    ax.set_ylabel("y_pred")
    plt.show()
    
    return model_rmse, model_mae, model_r2
```

```python
PR_ION = 'Cu' # 'Cu', 'Ni', 'Pb', 'Al', 'Co', 'Cr', 'NO3'

x = df_norm.drop(['Cu', 'Ni', 'Pb', 'Al', 'Co', 'Cr', 'NO3'], axis=1).to_numpy()
y = df_norm[[PR_ION]].to_numpy()

x.shape, y.shape
```

## 5. Scaling y-data
Train / Validation / Test = 0.7 / 0.2 / 0.1

TRAIN_TEST_RATIO = (Test) / (Train+Validation+Test)

VALIDATION_TRAIN_RATIO = Validation / (Train+Validation)

```python
TEST_TRAIN_RATIO = 0.1
VALIDATION_TRAIN_RATIO = 2 / 9 #7 / 9
RS = None

N_ITER_NO_CHANGE = 20
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=TEST_TRAIN_RATIO,
                                                    random_state=RS)
```

```python
scaler_y = MinMaxScaler()
y_train = scaler_y.fit_transform(y_train)
y_test = scaler_y.transform(y_test)

y_scaler_multiplier = scaler_y.data_max_[0] - scaler_y.data_min_[0]
y_scaler_multiplier
```

## 6. Model - MLP
Ordinary one-hidden-layer with *HIDDEN_LAYER_SIZES* neurons perceptron.
We use MLPRegressor from scikit-learn library.

```python
mlp_HIDDEN_LAYER_SIZES = [32]
mlp_ACTIVATION = 'tanh'
mlp_SOLVER = 'adam'
mlp_MAX_ITER = 10000
mlp_TOL = 1e-3
```

```python
model = MLPRegressor(hidden_layer_sizes=mlp_HIDDEN_LAYER_SIZES,
                     activation=mlp_ACTIVATION,
                     solver=mlp_SOLVER,
                     early_stopping=True,
                     validation_fraction=VALIDATION_TRAIN_RATIO,
                     n_iter_no_change=N_ITER_NO_CHANGE,
                     learning_rate_init = 0.001,
                     learning_rate='constant',
                     max_iter=mlp_MAX_ITER,
                     random_state=RS,
                     tol=mlp_TOL)
```

```python
model.get_params()
```

```python
model.fit(x_train, y_train.ravel())
```

```python
scatter_prediction(model, x_train, y_train, "Train data")
scatter_prediction(model, x_test, y_test, "Test data")
```

## 6.1 Cross-validation 1-hidden-layer MLP varying hidden_neurons.

```python
from sklearn.model_selection import train_test_split, KFold, cross_validate
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score, make_scorer

from sklearn.pipeline import Pipeline
from tqdm import tqdm

n_splits = 10
scoring = {'rmse': make_scorer(root_mean_squared_error), 
           'mae': make_scorer(mean_absolute_error)} # ['neg_root_mean_squared_error', 'neg_mean_absolute_error']

l_hn = [5, 7, 10, 15, 20, 32, 64] # list(range(1, 64))

mean_mae, std_mae = [], []
mean_rmse, std_rmse = [], []

for hn in tqdm(l_hn):
    
    

    model = MLPRegressor(hidden_layer_sizes=hn,
                         activation=mlp_ACTIVATION,
                         solver=mlp_SOLVER,
                         early_stopping=True,
                         validation_fraction=VALIDATION_TRAIN_RATIO,
                         n_iter_no_change=N_ITER_NO_CHANGE,
                         learning_rate_init = 0.001,
                         learning_rate='constant',
                         max_iter=mlp_MAX_ITER,
                         random_state=RS,
                         tol=mlp_TOL)
    
    #estimators = [('scale', MinMaxScaler()), ('mlp', model)]
    #pipe = Pipeline(estimators)
    
    # Scaling y before train_test split in cv only because we aprior know the distribution of ions.
    y_all_scaled = scaler_y.transform(y)
    cv_rf_kf = cross_validate(model, x, y_all_scaled.ravel(), cv=KFold(shuffle=True, n_splits=n_splits), scoring=scoring)
    
    mean_mae.append(np.mean(cv_rf_kf['test_mae'])) #test_neg_mean_absolute_error *(-1)
    std_mae.append(np.std(cv_rf_kf['test_mae'])) #test_neg_mean_absolute_error *(-1)
    
    mean_rmse.append(np.mean(cv_rf_kf['test_rmse'])) #test_neg_root_mean_squared_error *(-1)
    std_rmse.append(np.std(cv_rf_kf['test_rmse'])) #test_neg_root_mean_squared_error *(-1)
```

## 8. Model - KAN


```python
import torch
from kan import KAN

from tools import KAN_es, cross_val_KAN_es, kfold_experiment_es

def scatter_prediction_KAN(model, 
                       x, 
                       y_true, 
                       title = ""):
    """Evaluates MSE and R^2 of model. 
    Draws scatter plot of (y_true, y_predicted).
    Returns  (mse, r2)
    """
    y_pred = model.forward(x).detach().numpy()
    y_true = y_true.detach().numpy()
    
    model_rmse = rmse(y_true, y_pred)
    model_mae = mae(y_true, y_pred)
    model_r2 = r2_score(y_true, y_pred)
    
    fig, ax = plt.subplots(figsize=(5,5))
    plt.plot(y_true, y_pred, 'o', markersize=2, alpha = 0.2) 
    plt.plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], markersize=2, alpha = 0.5, color='r') 
    ax.set_title(f"RMSE={round(model_rmse, 5)} | MAE={round(model_mae, 5)} | R2={round(model_r2, 5)} | {title}")
    ax.set_xlabel("y_true")
    ax.set_ylabel("y_pred")
    plt.show()
    
    return model_rmse, model_mae, model_r2
```
